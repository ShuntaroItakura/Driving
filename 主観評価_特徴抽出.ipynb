```python
df5[df5.columns[10:13]] = 6 - df5[df5.columns[10:13]]
df5[df5.columns[10:13]]
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from sklearn.preprocessing import StandardScaler

# データの標準化
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df5)

# Autoencoderの定義
input_dim = data_scaled.shape[1]
encoding_dim = 3  # 次元圧縮後の次元数

input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu')(input_layer)
decoded = Dense(input_dim, activation='sigmoid')(encoded)

# Autoencoderモデル
autoencoder = Model(input_layer, decoded)

# Encoderモデル
encoder = Model(input_layer, encoded)

# Decoderモデル
encoded_input = Input(shape=(encoding_dim,))
decoder_layer = autoencoder.layers[-1]
decoder = Model(encoded_input, decoder_layer(encoded_input))

# モデルのコンパイル
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

# モデルの学習
autoencoder.fit(data_scaled, data_scaled,
                epochs=50,
                batch_size=32,
                shuffle=True,
                validation_split=0.2)

# データの次元圧縮
encoded_data = encoder.predict(data_scaled)

# 次元圧縮結果の表示
print(encoded_data)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# PCAで2次元に変換
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(encoded_data)

# 2次元プロット
plt.scatter(reduced_data[:, 0], reduced_data[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2D PCA of Encoded Data')
plt.show()
```

```python
#df5を主成分分析
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
#japanize_matplotlib
import japanize_matplotlib

import seaborn as sns

pca = PCA()
pca.fit(df1)

# データを主成分空間に写像
feature = pca.transform(df1)

#寄与率と累積寄与率の計算
ev_ratio = pca.explained_variance_ratio_
cum_ev_ratio = ev_ratio.cumsum()

#主成分負荷量を取得
loadings = pca.components_

#結果の表示
print("各主成分の寄与率:")
for i, ev in enumerate(ev_ratio):
    print(f"PC{i+1}: {ev:.4f}")

print("\n各主成分の累積寄与率:")
for i, cev in enumerate(cum_ev_ratio):
    print(f"PC{i+1}: {cev:.4f}")

# 主成分得点のプロット
plt.figure(figsize=(8, 6))



# 第一主成分と第二主成分でプロットする
plt.figure()
plt.scatter(feature[:, 0], feature[:, 1], alpha=0.8)
plt.title('Principal component analysis')
plt.xlabel('The first principal component')
plt.ylabel('The second principal component')
plt.grid()
plt.show()
```

```python
#df5を主成分分析 (主成分数3)
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

pca = PCA(n_components=3)
pca.fit(df5)

# データを主成分空間に写像
feature = pca.transform(df5)

#寄与率と累積寄与率の計算
ev_ratio = pca.explained_variance_ratio_
cum_ev_ratio = ev_ratio.cumsum()

#主成分負荷量を取得
loadings = pca.components_

#結果の表示
print("各主成分の寄与率:")
for i, ev in enumerate(ev_ratio):
    print(f"PC{i+1}: {ev:.4f}")

print("\n各主成分の累積寄与率:")
for i, cev in enumerate(cum_ev_ratio):
    print(f"PC{i+1}: {cev:.4f}")

# 主成分得点のプロット
plt.figure(figsize=(8, 6))

# 第一主成分と第二主成分でプロットする
plt.figure()
plt.scatter(feature[:, 0], feature[:, 1], alpha=0.8)
plt.title('Principal component analysis')
plt.xlabel('The first principal component')
plt.ylabel('The second principal component')
plt.grid()
plt.show()

#主成分負荷量のプロット
plt.figure()
plt.matshow(loadings, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(df5.columns)), df5.columns, rotation=90)
plt.show()


```
