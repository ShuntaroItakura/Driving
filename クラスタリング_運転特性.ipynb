```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_all = pd.read_csv('all.csv')
df_all
```

```python
df_all.columns
```

```python
df_all_in = df_all[['d1', 'd2', 'd3', 'd4', 'd5', 'w1', 'w2', 'w3', 'w4','w5', 'c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'p']]
df_all_in
```

```python
#クラスタリング
#最適なクラスタ数を求める
from sklearn.cluster import KMeans
dist_list = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, random_state=0)
    y_km = km.fit_predict(df_all_in)
    dist_list.append(km.inertia_)
plt.plot(range(1, 11), dist_list, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
#クラスタリング
km = KMeans(n_clusters=5, random_state=0)
y_km = km.fit_predict(df_all_in)
df_all_in['cluster'] = y_km
df_all_in

# #クラスタリング結果を可視化
# sns.pairplot(df_all_in, hue='cluster', diag_kind='kde')
# plt.show()

```

```python
#t-SNE
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
tsne.fit(df_all_in.iloc[:, :-1])
x = tsne.fit_transform(df_all_in.iloc[:, :-1])
df_all_in['x'] = x[:, 0]
df_all_in['y'] = x[:, 1]
sns.scatterplot(x='x', y='y', hue='cluster', data=df_all_in)
plt.show()
```

```python
#各データの平均値を求める
df_all_in.groupby('cluster').mean()
```

```python
df_all_s = df_all[['none_s', 'image_s','announce_s', 'image_sound_s', 'image_announce_s']]
df_all_s
```

```python
#df_all_inとdf_all_sを結合
df_all_s['cluster'] = y_km
df_all_s
```

```python
#各データの平均値を求める
df_all_s.groupby('cluster').mean()
```

```python
#階層的クラスタリング
from scipy.cluster.hierarchy import linkage, dendrogram
row_clusters = linkage(df_all_s, method='ward')
dendr = dendrogram(row_clusters)
plt.show()
```

```python
#階層的クラスタリング
#d1~d5でクラスタリング
df_all_d = df_all[['d1', 'd2', 'd3', 'd4', 'd5']]
row_clusters = linkage(df_all_d, method='ward')
dendr = dendrogram(row_clusters)
plt.show()

#各クラスターごとに平均値を求める
df_all_d['cluster'] = y_km
df_all_d.groupby('cluster').mean()

```
