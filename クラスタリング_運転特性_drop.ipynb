```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_all_drop_exp = pd.read_csv('all_drop_exp.csv')
df_all_drop_exp
```

```python
df_all_in = df_all_drop_exp[['d1', 'd2', 'd3', 'd4', 'd5', 'w1', 'w2', 'w3', 'w4','w5', 'c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'p']]
df_all_in.describe
```

```python
#クラスタリング
#最適なクラスタ数を求める
from sklearn.cluster import KMeans
dist_list = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, random_state=0)
    y_km = km.fit_predict(df_all_in)
    dist_list.append(km.inertia_)
plt.plot(range(1, 11), dist_list, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
#クラスタリング
km = KMeans(n_clusters=5, random_state=0)
y_km = km.fit_predict(df_all_in)
df_all_in['cluster'] = y_km
df_all_in

# #クラスタリング結果を可視化
# sns.pairplot(df_all_in, hue='cluster', diag_kind='kde')
# plt.show()

```

```python
#t-SNE
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
tsne.fit(df_all_in.iloc[:, :-1])
x = tsne.fit_transform(df_all_in.iloc[:, :-1])
df_all_in['x'] = x[:, 0]
df_all_in['y'] = x[:, 1]
sns.scatterplot(x='x', y='y', hue='cluster', data=df_all_in)
plt.savefig('img_tsne/tsne_kmeans.png')
plt.show()
```

```python
#各データの平均値を求める
df_all_in.groupby('cluster').mean()
```

```python
# #等高線を出力
# from matplotlib.colors import LogNorm
# from scipy.stats import multivariate_normal
# x = np.linspace(-10, 10, 100)
# y = np.linspace(-10, 10, 100)
# X, Y = np.meshgrid(x, y)
# XX = np.array([X.ravel(), Y.ravel()]).T
# Z = -gmm.score_samples(XX)
# Z = Z.reshape(X.shape)
# plt.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0), levels=np.logspace(0, 3, 10))
# plt.scatter(df_all_in['x'], df_all_in['y'], c=y_gmm)



import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.manifold import TSNE
from scipy.stats import gaussian_kde

# df_all_inをロードまたは定義する
# df_all_in = pd.read_csv('path_to_your_csv.csv')  # CSVからデータをロードする場合

# Gaussian Mixture Model
gmm = GaussianMixture(n_components=5, random_state=0)
y_gmm = gmm.fit_predict(df_all_in)
df_all_in['cluster'] = y_gmm

# t-SNE
tsne = TSNE(n_components=2, random_state=0)
x = tsne.fit_transform(df_all_in.iloc[:, :])
df_all_in['x'] = x[:, 0]
df_all_in['y'] = x[:, 1]

# クラスタリング結果と等高線をプロット
# クラスタリング結果
plt.figure(figsize=(10, 8))
sns.scatterplot(x='x', y='y', hue='cluster', data=df_all_in, palette='viridis', s=50)
plt.title('t-SNE with GMM Clusters')
plt.xlabel('t-SNE component 1')
plt.ylabel('t-SNE component 2')
plt.legend()
plt.savefig('img_tsne/tsne_gmm.png')
plt.show()

# カーネル密度推定による等高線プロット
plt.figure(figsize=(10, 8))
sns.scatterplot(x='x', y='y', hue='cluster', data=df_all_in, palette='viridis', s=50)

# KDE
kde = gaussian_kde(x.T)
xmin, ymin = x.min(axis=0) - 1
xmax, ymax = x.max(axis=0) + 1
xx, yy = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
positions = np.vstack([xx.ravel(), yy.ravel()])
f = np.reshape(kde(positions).T, xx.shape)

# 等高線をプロット
plt.contour(xx, yy, f, colors='k', linewidths=0.5)
plt.title('t-SNE with GMM Clusters and Contour Lines')
plt.xlabel('t-SNE component 1')
plt.ylabel('t-SNE component 2')
plt.legend()
plt.savefig('img_tsne/tsne_gmm_contour.png')
plt.show()
```
