semopy 使い方　https://note.com/cute_iris436/n/n54ce4cdd6724


```python
model1 = """
        s1 ~ d1
        s1 ~ d2
        s1 ~ d3
        d1 ~~ d2
        s1 ~ d4
        s1 ~ d5
        s1 ~ s2
        s1 ~ w1
        w5 ~ c2
        d3 ~ w5
        d2 ~ w2
        p ~ d1
        s1 ~ p
        s2 ~ p
        d1 ~~ d2    
       """

model = semopy.Model(model1)
model.fit(data=data1, obj='MLW')



ins = model.inspect(std_est=True)
print(ins)

```

```python
data1.columns
```

```python
model2 = """
    d1 ~ d1
    d2 ~ d2
    d3 ~ d3
    d4 ~ d4
    d5 ~ d5
    d1 ~~ d2    
    d1 ~~ d3
    d1 ~~ d5
    d2 ~~ d3
    d2 ~~ d4
    d2 ~~ d5
    d3 ~~ d5
    d4 ~~ d5
       """

model = semopy.Model(model2)
model.fit(data=data1, obj='MLW')



ins = model.inspect(std_est=True)
print(ins)
```

```python
semopy.semplot(model, "pd.png", plot_covs=True)
```

```python
#モデルの適合
stats = semopy.calc_stats(model)
print(stats.T)



#semopy.semplot(model1, "graph.png", plot_covs=True, engine="C:/Program Files/Graphviz/bin/dot.exe")
```
